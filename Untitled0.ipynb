```python
import pandas as pd
import json

# اسم الملف الذي رفعته (تأكد أن الاسم مطابق تماماً لما رفعته في القائمة الجانبية)
file_name = "Copy of Device ^0 Acc. Traceability.xlsm - version 35.0. 1.31.2026 3.xlsm"

try:
    # [cite_start]قراءة جميع أوراق العمل (من Sheet1 إلى Sheet7) [cite: 1, 2]
    all_sheets = pd.read_excel(file_name, sheet_name=None)

    # تحويل البيانات إلى تنسيق JSON مرن
    final_json = {}
    for name, df in all_sheets.items(): # هنا تم إضافة النقطتين : التي سببت الخطأ
        # [cite_start]تنظيف البيانات من الصفوف الفارغة تماماً [cite: 2]
        clean_df = df.dropna(how='all')
        # [cite_start]تحويل كل صف إلى قاموس (Dictionary) [cite: 3]
        final_json[name] = clean_df.to_dict(orient='records')

    # حفظ الملف الناتج بصيغة JSON
    with open('device_traceability.json', 'w', encoding='utf-8') as f:
        json.dump(final_json, f, ensure_ascii=False, indent=4)

    print("✅ تم استخراج البيانات بنجاح! الآن انظر للقائمة الجانبية وحمل الملف.")

except Exception as e:
    print(f"❌ حدث خطأ: {e}")
    print("تأكد أنك رفعت الملف بنفس الاسم المكتوب في الكود.")
```
